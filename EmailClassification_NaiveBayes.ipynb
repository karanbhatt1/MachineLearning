```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
vectorized_input_data = vectorizer.fit_transform(df['text'])
```

```python
vectorized_input_data.shape
```

```python
vectorcol = vectorizer.get_feature_names_out()
```

```python
vectorcol[5000:5100]
```

```python
df['spam'].value_counts() # check for imbalance data
```

```python
X = vectorized_input_data
y = df['spam']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y                  )
```

```python
base_model = MultinomialNB()
```

```python
mdl = base_model.fit(x_train,y_train)
```

```python
predictions = mdl.predict(x_test)
```

```python
Emails = ['We are Happy to tell that you are elligible for pay later program.Come Fast and acquire the advantage',
          "Hey, Lets have a quick meeting tommorow morning at 10 AM. Do join 5 min prior to meeting."]
```

```python
vecotremail = vectorizer.transform(Emails)
```

```python
y_test
```

```python
prediction = mdl.predict(vecotremail)
print(prediction)
for i in range(len(Emails)):
  if(prediction[i] == 1):
      print('Spam Mail')
  else:
      print('not spam Mail')
```

```python
roc_auc_score(y_test,mdl.predict(x_test))
```
